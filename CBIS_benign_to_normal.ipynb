### CBIS-DDSM 데이터의 benign 태그를 전부 normal태그로 변경하고 NCC 데이터와 합침

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/home/huray/data/CBIS-DDSM/img/data_whole_3class.csv', header=None)
df
```

```python
df[[1]] = df[[1]].astype(object)
df[[2]] = df[[2]].astype(object)
df[[3]] = df[[3]].astype(object)
df[[4]] = df[[4]].astype(object)
```

```python
for i, row in df.iterrows():
    if row[5] == 'B':
        df.iat[i,1] = ''
        df.iat[i,2] = ''
        df.iat[i,3] = ''
        df.iat[i,4] = ''
        df.iat[i,5] = ''
```

```python
df
```

```python
df.to_csv('/home/huray/data/CBIS-DDSM/img/data_whole_benign_is_normal.csv', header=False, index=False)
```
